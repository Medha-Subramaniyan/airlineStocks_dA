#2. **Daily returns by ticker**
    
    Create `ret_d = close.pct_change()` within each ticker (grouped). 
    Count **negative-return days vs positive-return days** per ticker.

```python
closes = [col for col in compiled_df.columns if col.startswith('Close_')]

#iterate through the columns of the dataframe
#when we land on a column that matches one of the closes values
    #create a new column called Ret_XXX that calculate pct_change() on the c olumn
```

```python
for col in closes:
    tick = col.split("_")
    col_name = 'ret_' + tick[1]
    compiled_df[col_name] = compiled_df[col].pct_change()
    
    
        
```

```python
#negative vs positive return days per ticker 

#create function that only does subsetting on ret columns 
    #create a dataframe with columns as tickers
    #rows (positive return days, negative return days)

rets = [col for col in compiled_df.columns if col.startswith('ret_')]
returns = pd.DataFrame(0, columns=['DAL', 'LUV', 'SAVE', 'UAL'], 
                      index=['positive', 'negative'])
```

```python
for col in rets: #iterate through the return columns
    splut = col.split('_') 
    ticker = splut[1]     #get the ticker 
    
    # Iterate through each row in the column to count positive/negative returns
    for value in compiled_df[col]: #extract the actual value 
        if value > 0:
            returns.loc['positive', ticker] = returns.loc['positive', ticker] + 1
        elif value < 0:
            returns.loc['negative', ticker] = returns.loc['negative', ticker] + 1

# Display the results
print(returns)
```

#3
Group by ticker + calendar week and compute: total dollar volume = sum(close * volume). Rank tickers by median weekly dollar volume.

```python
# Convert Date to datetime first
compiled_df['Date'] = pd.to_datetime(compiled_df['Date'])
ticker_data = []

# Extract data for each ticker
for ticker in ['DAL', 'LUV', 'SAVE', 'UAL']:
    # Get the close and volume columns for this ticker
    close_col = f'Close_{ticker}'
    volume_col = f'Volume_{ticker}'
    
    # Create a DataFrame for this ticker
    ticker_df = compiled_df[['Date', close_col, volume_col]].copy()
    ticker_df['ticker'] = ticker
    ticker_df['close'] = ticker_df[close_col]
    ticker_df['volume'] = ticker_df[volume_col]
    ticker_df = ticker_df[['Date', 'ticker', 'close', 'volume']]
    
    ticker_data.append(ticker_df)


all_tickers = pd.concat(ticker_data, ignore_index=True)

# Add calendar week
all_tickers['year'] = all_tickers['Date'].dt.year
all_tickers['week'] = all_tickers['Date'].dt.isocalendar().week

# Calculate daily dollar volume
all_tickers['daily_dollar_volume'] = all_tickers['close'] * all_tickers['volume']

# Group by ticker and calendar week, sum the daily dollar volumes
weekly_dollar_volume = all_tickers.groupby(['ticker', 'year', 'week'])['daily_dollar_volume'].sum().reset_index()
weekly_dollar_volume.columns = ['ticker', 'year', 'week', 'weekly_dollar_volume']

# Calculate median weekly dollar volume for each ticker
median_weekly_dv = weekly_dollar_volume.groupby('ticker')['weekly_dollar_volume'].median().reset_index()
median_weekly_dv.columns = ['ticker', 'median_weekly_dollar_volume']

# Rank tickers by median weekly dollar volume (highest to lowest)
median_weekly_dv_ranked = median_weekly_dv.sort_values('median_weekly_dollar_volume', ascending=False).reset_index(drop=True)

median_weekly_dv_ranked
```

#4 Group by ticker + year-month to compute: first open, max high, min low, last close. Output a table for 2019 only.

```python
# Convert Date to datetime if not already done
compiled_df['Date'] = pd.to_datetime(compiled_df['Date'])
# Create a list to store individual ticker data
ticker_data = []

# Extract data for each ticker
for ticker in ['DAL', 'LUV', 'SAVE', 'UAL']:
    # Get the OHLC columns for this ticker
    open_col = f'Open_{ticker}'
    high_col = f'High_{ticker}'
    low_col = f'Low_{ticker}'
    close_col = f'Close_{ticker}'
    
    # Create a DataFrame for this ticker
    ticker_df = compiled_df[['Date', open_col, high_col, low_col, close_col]].copy()
    ticker_df['ticker'] = ticker
    ticker_df['open'] = ticker_df[open_col]
    ticker_df['high'] = ticker_df[high_col]
    ticker_df['low'] = ticker_df[low_col]
    ticker_df['close'] = ticker_df[close_col]
    ticker_df = ticker_df[['Date', 'ticker', 'open', 'high', 'low', 'close']]
    
    ticker_data.append(ticker_df)

# Combine all ticker data
all_tickers = pd.concat(ticker_data, ignore_index=True)

# Add year-month column
all_tickers['year_month'] = all_tickers['Date'].dt.to_period('M')

# Filter for 2019 only
tickers_2019 = all_tickers[all_tickers['Date'].dt.year == 2019]

# Group by ticker and year-month, compute OHLC statistics
monthly_ohlc = tickers_2019.groupby(['ticker', 'year_month']).agg({
    'open': 'first',    # First open of the month
    'high': 'max',      # Maximum high of the month
    'low': 'min',       # Minimum low of the month
    'close': 'last'     # Last close of the month
}).reset_index()

monthly_ohlc.columns = ['ticker', 'year_month', 'first_open', 'max_high', 'min_low', 'last_close']

# Sort by ticker and year_month for better readability
monthly_ohlc = monthly_ohlc.sort_values(['ticker', 'year_month'])
monthly_ohlc
```

#5 Within each ticker, compute 20-day rolling std of daily returns. For each month, identify the top 2 most volatile tickers (aggregate by month on the rolling std).

```python
# Convert Date to datetime if not already done
compiled_df['Date'] = pd.to_datetime(compiled_df['Date'])

# Create a list to store individual ticker data with returns
ticker_data = []

# Extract data for each ticker
for ticker in ['DAL', 'LUV', 'SAVE', 'UAL']:
    # Get the return column for this ticker
    ret_col = f'ret_{ticker}'
    
    # Create a DataFrame for this ticker
    ticker_df = compiled_df[['Date', ret_col]].copy()
    ticker_df['ticker'] = ticker
    ticker_df['returns'] = ticker_df[ret_col]
    ticker_df = ticker_df[['Date', 'ticker', 'returns']]
    
    ticker_data.append(ticker_df)

# Combine all ticker data
all_tickers = pd.concat(ticker_data, ignore_index=True)

# Add year-month column
all_tickers['year_month'] = all_tickers['Date'].dt.to_period('M')

# Sort by ticker and date for rolling calculations
all_tickers = all_tickers.sort_values(['ticker', 'Date'])

# Compute 20-day rolling standard deviation for each ticker
all_tickers['rolling_std_20d'] = all_tickers.groupby('ticker')['returns'].rolling(
    window=20, min_periods=1
).std().reset_index(0, drop=True)

# Group by ticker and year-month, compute mean of rolling std for each month
monthly_volatility = all_tickers.groupby(['ticker', 'year_month'])['rolling_std_20d'].mean().reset_index()
monthly_volatility.columns = ['ticker', 'year_month', 'avg_rolling_std_20d']

# For each month, identify the top 2 most volatile tickers
top_volatile_tickers = monthly_volatility.groupby('year_month').apply(
    lambda x: x.nlargest(2, 'avg_rolling_std_20d')
).reset_index(drop=True)

# Sort by year_month and volatility (descending)
top_volatile_tickers = top_volatile_tickers.sort_values(['year_month', 'avg_rolling_std_20d'], ascending=[True, False])

top_volatile_tickers
```
